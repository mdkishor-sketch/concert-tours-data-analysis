```python
#STRUCTURE AND OVERVIEW
import pandas as pd

df = pd.read_csv(os.path.join(dataset_path, os.listdir(dataset_path)[0]))

df.head()
df.tail()
df.sample()
df.shape
df.info()
df.describe()
df.describe(include = "object")
```

```python
df.columns = (df.columns
              .str.strip()                           # Remove whitespace
              .str.lower()                           # Convert to lowercase
              .str.replace('_', '') )                # Replace spaces

df = df.rename(columns={
    'actualgross': 'actual_gross',
    'adjustedgross2022': 'adjusted_gross_2022',
    'alltimepeak': 'all_time_peak',
    'tourtitle' : 'tour_title',
    'averagegross' : 'average_gross'
})

df.head()
```

```python
#Data(Value) Modification
money_cols = ["actual_gross","adjusted_gross_2022","average_gross"]

for col in money_cols:
    df[col] = (
        df[col]
        .replace(r"[\$,]", "", regex=True)
    )

df["all_time_peak"] = (
    df["all_time_peak"]
    .astype(str)
    .str.extract(r"(\d+)")
)

df["years"] = df["years"].str.replace("–", "-", regex=False)

df["start_year"] = df["years"].str.split("-").str[0].astype(int)
df["end_year"] = (
    df["years"].str.split("-").str[1]
    .fillna(df["start_year"])
    .astype(int)
)

df["tour_title"] = df["tour_title"].str.replace(r"\[.*?\]", "", regex=True).str.strip()
df.drop(columns="ref", inplace=True)

df.head()
```
